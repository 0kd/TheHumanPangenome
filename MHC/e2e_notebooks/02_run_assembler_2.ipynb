```python
%%bash
mkdir -p  /home/dnanexus/phased_asm_2/
cd /home/dnanexus/phased_asm_2/
dx download --recursive -f /phased_reads_2/
```

```python
%%bash
cd /home/dnanexus/phased_asm_2/
git clone https://github.com/lh3/seqtk.git
cd seqtk
make
```

```python
%%bash
cd /home/dnanexus/phased_asm_2/

zcat phased_reads_2/HG002.Sequel.15kb.pbmm2.hs37d5.whatshap.haplotag.RTG.10x.trio.MHConly.fastq.gz | awk 'NR % 4 ==1' | tr -d "@" | sort > ids1
cat phased_reads_2/asm_mapped.fa | grep ">" | tr -d ">" | sort > ids2
comm  -13 ids1 ids2  > unphased_reads_on_assembly_ids
./seqtk/seqtk subseq phased_reads_2/asm_mapped.fa unphased_reads_on_assembly_ids > unphased_reads_on_assembly.fa
```

```python
%%bash
cd /home/dnanexus/phased_asm_2/
echo /wd/phased_reads_2/HG002.Sequel.15kb.pbmm2.hs37d5.whatshap.haplotag.RTG.10x.trio.MHConly.28498559.H1.fastq.gz > reads.lst
echo /wd/phased_reads_2/HG002.Sequel.15kb.pbmm2.hs37d5.whatshap.haplotag.RTG.10x.trio.MHConly.28498559.untagged.fastq.gz >> reads.lst
echo /wd/unphased_reads_on_assembly.fa >> reads.lst

docker run --rm -v $PWD:/wd --user $(id -u):$(id -g) \
        --entrypoint /opt/entry_dev.sh cschin/peregrine:0.1.5.2 \
            asm /wd/reads.lst 4 4 4 4 4 4 4 4 4 --shimmer-r 3 \
            --with-consensus \
            --best_n_ovlp 8 \
            --output /wd/asm-HG002-MHC-H1-r3.pg0.1.5.2/
            
cp reads.lst asm-HG002-MHC-H1-r3.pg0.1.5.2/
```
